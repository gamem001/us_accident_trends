```python
import pandas as pd
import csv
import json
import numpy as np
from census import Census
import requests
from us import states
from api_key import weather_key
import matplotlib.pyplot as plt
```

```python
file_path = 'final_accident.csv'

final_accident_df = pd.read_csv(file_path)
final_accident_df.head()
```

```python
spring = final_accident_df[(final_accident_df['Date'] >= '2019-03-01') & (final_accident_df['Date'] <= '2019-05-31')]
spring.to_csv('spring_accidents.csv', index=False)
spring.head()
```

```python
summer = final_accident_df[(final_accident_df['Date'] >= '2019-06-01') & (final_accident_df['Date'] <= '2019-08-31')]
summer.to_csv('summer_accidents.csv', index=False)
summer.head()
```

```python
fall = final_accident_df[(final_accident_df['Date'] >= '2019-09-01') & (final_accident_df['Date'] <= '2019-11-30')]
fall.to_csv('fall_accidents.csv', index=False)
fall.head()
```

```python
winter_1 = final_accident_df[(final_accident_df['Date'] >= '2019-12-01') & (final_accident_df['Date'] <= '2019-12-31')]
winter_1

winter_2 = final_accident_df[(final_accident_df['Date'] >= '2019-01-01') & (final_accident_df['Date'] <= '2019-02-28')]
winter_2

winter = winter_1.append(winter_2)
winter.to_csv('winter_accidents.csv', index=False)
winter
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
label = ['Spring', 'Summer', 'Fall', 'Winter']
totals = [spring, summer, fall, winter]
colors = ['green', 'yellow', 'orange', 'lightskyblue']
plt.title('Accidents Per Season United States')
plt.xlabel('Season')
plt.ylabel('Num of Accidents')
ax.bar(label,totals, color = colors)
plt.show()
```
